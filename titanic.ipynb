```python
import pandas as pd
```

```python
titanic_data = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
titanic_data.head()
```

```python
print("Shape:", titanic_data.shape)
```

```python
print(titanic_data.dtypes)
```

```python
print(titanic_data.isnull().sum())
```

```python
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
```

```python
print(titanic_data['Survived'].value_counts())
```

```python
print(titanic_data.groupby('Sex')['Survived'].mean() * 100)
```

```python
print(titanic_data.groupby('Survived')['Age'].mean())
```

```python
print(titanic_data['Embarked'].mode()[0])
```

```python

print(titanic_data['Pclass'].value_counts())
```

```python

print(titanic_data[titanic_data['Fare'] > 100])
```

```python

max_fare = titanic_data['Fare'].max()
print("Max Fare:", max_fare)
print(titanic_data[titanic_data['Fare'] == max_fare])
```

```python

print(titanic_data.groupby('Pclass')['Fare'].mean())
```

```python

titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch'] + 1
print(titanic_data[['FamilySize']].head(10))

```

```python

print(titanic_data.sort_values(by='Age', ascending=False).head(5))
```
