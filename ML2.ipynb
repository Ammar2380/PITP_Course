```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np

data = {
    "Marks": [30, 40, 50, 60, 70, 80, 90],
    "Result": [0, 0, 0, 1, 1, 1, 1],
}

df = pd.DataFrame(data)

x = df[["Marks"]]
y = df["Result"]

clf = LogisticRegression()
clf.fit(x, y)
pre = clf.predict([[30]])
if pre[0] == 1:
    print(True)
else:
    print(False)
```
