```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


X = np.array([50, 100, 150, 200, 250]).reshape(-1, 1)
y = np.array([500, 1000, 1500, 2000, 2500])

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)


plt.scatter(X, y)
plt.plot(X, y_pred)
plt.xlabel("Units Consumed")
plt.ylabel("Electricity Bill")
plt.title("Electricity Bill Prediction")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

from mpl_toolkits.mplot3d import Axes3D

# Data
size = np.array([500, 800, 1000, 1200, 1500])
bedrooms = np.array([1, 2, 2, 3, 3])
rent = np.array([10000, 15000, 18000, 22000, 27000])

X = np.column_stack((size, bedrooms))

model = LinearRegression()
model.fit(X, rent)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(size, bedrooms, rent)

ax.set_xlabel("Size")
ax.set_ylabel("Bedrooms")
ax.set_zlabel("Rent")

plt.title("House Rent Prediction")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

from mpl_toolkits.mplot3d import Axes3D

mileage = np.array([15, 18, 20, 22, 25])
year = np.array([2015, 2017, 2018, 2020, 2022])
price = np.array([800000, 900000, 1000000, 1200000, 1500000])

X = np.column_stack((mileage, year))

model = LinearRegression()
model.fit(X, price)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(mileage, year, price)

ax.set_xlabel("Mileage")
ax.set_ylabel("Year")
ax.set_zlabel("Price")

plt.title("Car Price Prediction")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

from mpl_toolkits.mplot3d import Axes3D

rainfall = np.array([100, 150, 200, 250, 300])
fertilizer = np.array([50, 60, 70, 80, 90])
yield_ = np.array([20, 30, 40, 50, 60])

X = np.column_stack((rainfall, fertilizer))

model = LinearRegression()
model.fit(X, yield_)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(rainfall, fertilizer, yield_)

ax.set_xlabel("Rainfall")
ax.set_ylabel("Fertilizer")
ax.set_zlabel("Yield")

plt.title("Crop Yield Prediction")
plt.show()
```
